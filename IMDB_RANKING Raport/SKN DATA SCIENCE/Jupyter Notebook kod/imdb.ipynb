```python
df2 = pd.read_csv('imdb_cleaned.csv',sep =',')
df2.head()
```

```python
df2.dtypes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
ranking = df2['imdb_rating']

min_ranking = ranking.min()
max_ranking = ranking.max()
mean_ranking = ranking.mean()



plt.figure(figsize=(15,10))
sns.histplot(ranking, bins=20, kde=True, color='darkgreen', stat='count', alpha=0.5)


plt.axvline(min_ranking, color='blue', linestyle='--', linewidth=2)
plt.axvline(max_ranking, color='red', linestyle='--', linewidth=2)
plt.axvline(mean_ranking, color='purple', linestyle='--', linewidth=2)

plt.text(min_ranking, plt.ylim()[1] * -0.07, f'Min: {min_ranking:.2f}', color='blue', fontsize=12, ha='center')
plt.text(max_ranking, plt.ylim()[1] * -0.07, f'Max: {max_ranking:.2f}', color='red', fontsize=12, ha='center')
plt.text(mean_ranking, plt.ylim()[1] * -0.07, f'Średnia: {mean_ranking:.2f}', color='purple', fontsize=12, ha='center')

plt.title('Rozkład ocen IMDB', fontsize=16)
plt.xlabel('Oceny IMDB', fontsize=12)
plt.ylabel('Ilość filmów', fontsize=12)


plt.savefig('Rozkład ocen IMDB.png') 
plt.show()
```

```python
reżyser = df2['director'].value_counts().nlargest(12).index

plt.figure(figsize=(15, 10))
sns.boxplot(x='director', y='imdb_rating', data=df2[df2['director'].isin(reżyser)])
plt.xticks(rotation=45)

plt.axhline(mean_ranking, color='purple', linestyle='--', linewidth=2)
plt.text(x=-0,y=mean_ranking ,s=f'Średnia ocena: {mean_ranking:.2f}',  color='purple', fontsize=12, ha='right',rotation=30,)
plt.title('Reżyserzy a ocena ich filmów', fontsize=16)

plt.xlabel('Reżyser')
plt.ylabel('Ocena IMDB')
plt.savefig('Reżyserzy a ocena ich filmów.png') 
plt.show()
```

```python
df4 = pd.read_csv('imdb_gatunki.csv', header=None)
df5 = df4[0].str.split(',', expand=True)
df5 = df5[df5[0] != 'genre'] 


df5.columns = ['naz1', 'naz2', 'naz3']

df6= df5.melt(value_name='gatunki')
```

```python

df6.drop('variable', axis=1)
df7=df6.dropna()
df7.head(30)
df8=df7.drop('variable', axis=1)
df8.head(30)
df8.isnull().sum()
```

```python
df8.value_counts()
```

```python
df9 = df8['gatunki'].str.strip()
df9.value_counts()
```

```python
df9 = pd.DataFrame(df9, columns=['gatunki'])

plt.figure(figsize=(12,8))
największe = df9['gatunki'].value_counts().index
sns.countplot(y='gatunki', data=df9, order=największe)
plt.title('Liczba wystąpień gatunku w 1000 filmach', fontsize=16)
plt.xlabel('Gatunek', fontsize=12)
plt.ylabel('Liczba wystąpień', fontsize=12)
plt.savefig('Liczba wystąpień gatunku w 1000 filmach.png') 
plt.show()
```
